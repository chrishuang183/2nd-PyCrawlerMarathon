```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
#response

import json
data = json.loads(response)
print("total title:",len(data))
data
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)

data
for d in data:    
    print("title:",d['title'],"Creat time:",d['mediaMeta'][0]['createdAt'], 
          "commentCount:",d['commentCount'],"likeCount:",d['likeCount']) 
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
f = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = r.text
response_f = f.text

data = json.loads(response)
data_f = json.loads(response_f)

print("total title with popular:",len(data))
print("total title with no popular:",len(data_f))

#data
print("########### popular ###########")
commentCount = 0
likeCount = 0
for d in data:    
    commentCount = commentCount + d['commentCount']
    likeCount = likeCount + d['likeCount']

print("total commentCount :", commentCount)
print("total likeCount :", likeCount)

print("########### No popular ###########")
#print(data_f)
commentCount = 0
likeCount = 0
for d in data_f:    
    commentCount = commentCount + d['commentCount']
    likeCount = likeCount + d['likeCount']

print("total commentCount :", commentCount)
print("total likeCount :", likeCount)
```
